# Matplotlib

Matplotlib is a popular Python library used for data visualization.
It helps you create graphs, charts, and plots to understand data visually.

# What is Matplotlib?

1. A plotting library for Python
2. Used in data analysis, machine learning, and reporting
3. Works well with libraries like NumPy and Pandas

# Why use Matplotlib?
Convert raw data → visual insights 📊
Easy to create:

    1. Line charts
    2. Bar charts
    3. Pie charts
    4. Histograms
    5. Scatter plots

# Types of Plots
# 1. Line Plot

A line plot shows data points connected by straight lines.

It is mainly used to show trends or changes over time

```python
plt.plot(x, y, 'r*--', linewidth=1) # color = 'red', linestyle='--', marker='*'

plt.show()
```

```python
x = [1, 2, 3]
y = [4, 5, 6]

plt.plot(x, y, color='blue', linestyle='--', marker='o', linewidth=2, label='Data')
plt.title("Sample Graph")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.legend(loc='upper left') # 'lower right'
plt.grid()

plt.show()
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
z = [7, 8, 9]

plt.plot(x, y, color='red', linestyle='--', marker='o', label='Sales')
plt.plot(x, z, color='blue', linestyle='-', marker='s', label='Profit')

plt.title("Company Report")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.grid()

plt.show()
```

# 2. Bar Chart

A bar chart represents data using rectangular bars, where:

1. Height (or length) of each bar = value
2. Each bar = one category

Used to compare values between categories

```python
x = [1, 2, 3]
y = [4, 5, 6]

plt.bar(x, y) # width=0.5, color = ["green", "black", "red"]

plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['A', 'B', 'C']
y = [10, 20, 15]

plt.bar(x, y, color='orange', edgecolor='black', label='Data')
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")

for i in range(len(x)):
    plt.text(x[i], y[i], y[i])

plt.legend()
plt.grid(axis='y')

plt.show()
```

# 3. Scatter Plot

A scatter plot is used to show the relationship between two variables by plotting points on a graph.

# What is a Scatter Plot?

1. Each point represents (x, y)
2. Shows how one variable affects another

# Example:

1. Height vs Weight
2. Study hours vs Marks

```python
plt.scatter(x, y) # color='red', alpha=0.5
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [4, 5, 6]

plt.scatter(x, y, color='green', s=100, marker='o', edgecolors='black', label='Points')
'''
sizes = [50, 100, 200]
plt.scatter(x, y, s=sizes)
'''
plt.title("Scatter Example")
plt.xlabel("X")
plt.ylabel("Y")

for i in range(len(x)):
    plt.text(x[i], y[i] + 0.2, y[i], ha='center')

'''
+ 0.2 → moves text slightly above
ha='center' → aligns text properly
'''

plt.legend()

plt.show()
```

# 4. Pie Chart

A pie chart is used to show how a whole is divided into parts (percentages). Each slice represents a category’s contribution.

# When to Use Pie Charts

1. You want to show percentage/share

# Examples:

1. Market share
2. Budget distribution
3. Student grade percentages

```python
import matplotlib.pyplot as plt

values = [30, 40, 30]
labels = ['A', 'B', 'C']

plt.pie(values, labels=labels)
plt.show()
```

```python
import matplotlib.pyplot as plt

values = [50, 30, 20]
labels = ['Food', 'Rent', 'Savings']

plt.pie(
    values,
    labels=labels,
    autopct='%1.1f%%',
    colors=['gold', 'lightblue', 'lightgreen'],
    explode=[0, 0.1, 0.2],
    startangle=90, # Controls where the pie chart starts (rotation) startangle=90 → starts from top (12 o’clock position)
    shadow=True
)

plt.title("Monthly Expenses")
plt.show()
```

# 5. Histogram

A Histogram in Matplotlib is used to show the distribution of numerical data (how data is spread across ranges).

# What is a Histogram?

1. It groups data into intervals (bins)
2. Shows frequency (count) of values in each range

Example: Marks, heights, ages, etc.

# Important Points

1. Works only with numeric data
2. Not for categories (use bar chart instead)

# Histogram vs Bar Chart

| Feature   | Histogram            | Bar Chart  |
| --------- | -------------------- | ---------- |
| Data type | Continuous (numbers) | Categories |
| Bars      | Touch each other     | Have gaps  |
| Use       | Distribution         | Comparison |


```python
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 60, 65, 70, 75, 80, 85, 85, 85] # Repeated Values So The Sizes are Increase 

plt.hist(marks, edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Frequency")

plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [45, 50, 55, 60, 60, 65, 70, 75, 80, 85, 85, 85]

plt.hist(marks, bins=5, color='green', edgecolor='black')
plt.title("Marks Distribution")
plt.xlabel("Marks Range")
plt.ylabel("Frequency")

plt.show()
```

# Matplotlib

# What are Methods in Matplotlib?

Methods = Functions you call using plt.

They define what type of chart or action you want

# Common Methods

| Method          | Purpose       |
| --------------- | ------------- |
| `plt.plot()`    | Line chart    |
| `plt.bar()`     | Bar chart     |
| `plt.scatter()` | Scatter plot  |
| `plt.hist()`    | Histogram     |
| `plt.pie()`     | Pie chart     |
| `plt.title()`   | Add title     |
| `plt.xlabel()`  | X-axis label  |
| `plt.ylabel()`  | Y-axis label  |
| `plt.legend()`  | Show legend   |
| `plt.grid()`    | Show grid     |
| `plt.show()`    | Display chart |
| `plt.savefig()` | Save chart    |



# What are Arguments (Parameters)?

Arguments = Values passed inside methods

They control how the chart looks

# A. Line Plot Arguments

| Argument    | Use         |
| ----------- | ----------- |
| `color`     | Line color  |
| `linestyle` | Line type   |
| `marker`    | Point shape |
| `linewidth` | Thickness   |
| `label`     | Legend name |


# B. Bar Chart Arguments

| Argument | Use       |
| -------- | --------- |
| `color`  | Bar color |
| `width`  | Bar width |


# C. Scatter Plot Arguments

| Argument | Use         |
| -------- | ----------- |
| `color`  | Point color |
| `s`      | Size        |


# D. Histogram Arguments

| Argument    | Use              |
| ----------- | ---------------- |
| `bins`      | Number of groups |
| `color`     | Bar color        |
| `edgecolor` | Border color     |


# E. Pie Chart Arguments

| Argument  | Use             |
| --------- | --------------- |
| `labels`  | Category names  |
| `autopct` | Percentage      |
| `explode` | Highlight slice |
