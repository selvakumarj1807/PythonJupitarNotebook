# Seaborn

Seaborn is a Python library used to create beautiful and easy graphs for data analysis.

It works on top of Matplotlib and makes charts look more attractive with less code.

# Simple Definition

Seaborn helps you visualize data using graphs and charts to understand patterns, trends, and relationships.

# Core Idea
1. Matplotlib is the base library → gives full control but requires more code
2. Seaborn is built on top of it → makes graphs easier, cleaner, and more attractive

# Think like:

1. Matplotlib = raw engine ⚙️
2. Seaborn = stylish UI 🚀

# Why Use Seaborn?
1. Easy to create charts
2. Attractive default styles
3. Good for data analytics and statistics
4. Works well with Pandas DataFrames

# Common Charts in Seaborn
1. Line Plot – Shows trends over time
2. Bar Plot – Compares categories
3. Histogram – Shows data distribution
4. Scatter Plot – Shows relationship between two variables
5. Heatmap – Displays correlation using colors

# 1. Line Plot

# Purpose

Used to show trends or changes over time.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [100, 150, 200, 180, 250]
}

df = pd.DataFrame(data)

sns.lineplot(x='Month', y='Sales', data=df)

plt.title("Monthly Sales Trend")
plt.show()
```

# 2. Bar Plot
# Purpose

Used to compare categories.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Department': ['IT', 'HR', 'Sales', 'Marketing'],
    'Employees': [50, 20, 35, 25]
}

df = pd.DataFrame(data)

sns.barplot(x='Department', y='Employees', data=df)

plt.title("Employees by Department")
plt.show()
```

# 3. Histogram
# Purpose

Used to show data distribution.

```python
import seaborn as sns
import matplotlib.pyplot as plt

numbers = [10, 20, 20, 30, 30, 30, 40, 50, 60]

sns.histplot(numbers)

plt.title("Histogram Example")
plt.show()
```

# 4. Scatter Plot
# Purpose

Used to show relationship between two variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Study_Hours': [1, 2, 3, 4, 5, 6, 3],
    'Marks': [35, 45, 60, 70, 90, 80, 50]
}

df = pd.DataFrame(data)

sns.scatterplot(x='Study_Hours', y='Marks', data=df)

plt.title("Study Hours vs Marks")
plt.show()
```

# 5. Heatmap
# Purpose

Used to show correlation using colors.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Math': [80, 70, 90, 60],
    'Science': [85, 75, 95, 65],
    'English': [78, 68, 88, 58]
}

df = pd.DataFrame(data)

correlation = df.corr()

sns.heatmap(correlation, annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

# Heatmap

# Math vs Science correlation

Math     = [80, 70, 90, 60]

Science  = [85, 75, 95, 65]

# To Find Step 1: Mean (Average)

Math mean:

x̄ = (80 + 70 + 90 + 60) / 4 = 75

Science mean:

ȳ = (85 + 75 + 95 + 65) / 4 = 80

# Step 2: Deviation (Value - Mean)

Math:

| Value | Mean | Deviation |
| ----- | ---- | --------- |
| 80    | 75   | +5        |
| 70    | 75   | -5        |
| 90    | 75   | +15       |
| 60    | 75   | -15       |


Science:

| Value | Mean | Deviation |
| ----- | ---- | --------- |
| 85    | 80   | +5        |
| 75    | 80   | -5        |
| 95    | 80   | +15       |
| 65    | 80   | -15       |

# Step 3: Multiply deviations

| Math Dev | Science Dev | Multiply |
| -------- | ----------- | -------- |
| 5        | 5           | 25       |
| -5       | -5          | 25       |
| 15       | 15          | 225      |
| -15      | -15         | 225      |

Sum of Multiply = 25 + 25 + 225 + 225 = 500

# Step 4: Square deviations

Math:

5^2+(−5)^2+15^2+(−15)^2 = 25+25+225+225 = 500

Science:

Same values → 500

# Step 5: Final Formula

| x  | y  | x - x̄ | y - ȳ | (x-x̄)(y-ȳ) | (x-x̄)² | (y-ȳ)² |
| -- | -- | ------ | ----- | ----------- | ------- | ------ |
| 80 | 85 | 5      | 5     | 25          | 25      | 25     |
| 70 | 75 | -5     | -5    | 25          | 25      | 25     |
| 90 | 95 | 15     | 15    | 225         | 225     | 225    |
| 60 | 65 | -15    | -15   | 225         | 225     | 225    |

Sum Values:

Σ(x-x̄)(y-ȳ) = 25 + 25 + 225 + 225 = 500

Σ(x-x̄)² = 25 + 25 + 225 + 225 = 500

Σ(y-ȳ)² = 25 + 25 + 225 + 225 = 500

Apply Formula:

r = Σ(x-x̄)(y-ȳ) / √ Σ(x-x̄)² Σ(y-ȳ)² 

r = 500 / √(500 × 500)

r = 500 / 500

r = 1
