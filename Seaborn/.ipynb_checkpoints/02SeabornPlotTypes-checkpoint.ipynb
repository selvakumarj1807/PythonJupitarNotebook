# Seaborn Plot Types

# What is Seaborn?

1. Seaborn is a Python visualization library used to create attractive and informative graphs.

2. It works on top of Matplotlib and is mainly used in Data Analytics and Data Science.

# Install Required Libraries

pip install seaborn matplotlib pandas

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# Load Sample Dataset

Seaborn provides built-in datasets.

```python
tips = sns.load_dataset("tips")

tips
```

# 1 Distribution Plots

Distribution plots show how data values are spread.

# 1.1. Histplot

# Purpose

Used to understand data distribution.

# Explanation
1. Shows how values are distributed
2. Helps identify frequency of data

```python
sns.histplot(tips['total_bill'])
plt.show()
```

# 1.2 KDE Plot
# Purpose

Shows smooth distribution curve.

# Explanation
1. Smooth line representation
2. Shows data density

```python
sns.kdeplot(tips['total_bill'])
plt.show()
```

# 1.3 Rug Plot
# Purpose

Shows individual data points.

# Explanation
1. Displays small lines for each value
2. Useful for understanding point distribution

```python
sns.rugplot(tips['total_bill'])
plt.show()
```

# 1.4 Displot
# Purpose

Combined histogram and KDE.

# Explanation
1. Advanced distribution plot
2. Can include histogram and KDE

```python
sns.displot(tips['total_bill'])
plt.show()
```

# 2. Categorical Data Plots

Used when working with category-based data.

# 2.1 Bar Plot
# Purpose

Compares average values between categories.

# Explanation
1. Compares categories
2. Example: Average bill by day

```python
sns.barplot(x='day', y='total_bill', data=tips)
plt.show()
```

# 2.2 Count Plot
# Purpose

Counts occurrences of categories.

# Explanation
1. Counts category frequency
2. Shows number of entries per day

```python
sns.countplot(x='day', data=tips)
plt.show()
```

# 2.3 Box Plot
# Purpose

Shows data spread and outliers.

# Explanation
1. Shows median
2. Shows minimum and maximum
3. Identifies outliers

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

# 2.4 Violin Plot
# Purpose

Shows distribution and density.

# Explanation
1. Similar to boxplot
2. Better distribution view

```python
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

# 2.5 Strip Plot
# Purpose

Shows all data points.

# Explanation
1. Shows individual points
2. Useful for small datasets

```python
sns.stripplot(x='day', y='total_bill', data=tips)
plt.show()
```

# 2.6 Swarm Plot
# Purpose

Avoids overlapping points.

# Explanation
1. Better version of stripplot
2. Prevents overlap

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```

# 3. Matrix Plots

Used to show relationships between variables.

# 3.1 Heatmap
# Purpose

Displays correlation using colors.

# Explanation
1. Shows relationships between columns
2. Color intensity indicates correlation strength

```python
corr = tips.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

# 3.2 Clustermap
# Purpose

Groups similar values.

# Explanation
1. Creates clustered heatmap
2. Finds similar patterns

# The outer black lines in your Seaborn clustermap() are called dendrograms.

They are not borders.

# They show:

1. how data is grouped
2. similarity between columns and rows

# Pre Requirements

pip install scipy

```python
!pip install scipy
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

corr = tips.corr(numeric_only=True)

sns.clustermap(corr, annot=True, linewidths=5, figsize=(8, 8), cmap='coolwarm')

# Title
plt.suptitle("Clustering", y=1.05)

plt.savefig('a.png')

plt.show()
```

# 4. Grids

Used to create multiple plots together.

# 4.1 Pair Plot
# Purpose

Shows relationship between all numeric columns.

# Explanation
1. Creates multiple scatter plots
2. Useful for feature analysis

```python
tips = sns.load_dataset("tips")
sns.pairplot(tips)
plt.show()
```

# 4.2 Facet Grid
# Purpose

Creates subplots based on categories.

# Explanation
1. Creates separate charts
2. Useful for comparison

```python
g = sns.FacetGrid(tips, col='time')
g.map(sns.histplot, 'total_bill')
plt.show()
```

# 5. Regression Plots

Used to understand relationship between variables.

# 5.1 Regression Plot
# Purpose

Shows trend line.

# Explanation
1. Scatter plot with regression line
2. Shows data trend

```python
sns.regplot(x='total_bill', y='tip', data=tips)
plt.show()
```

# 5.2 LM Plot
# Purpose

Regression plot with category separation.

# Explanation
1. Advanced regression graph
2. Useful for grouped data

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

# 6. Style and Color

Used to improve graph appearance.

# 6.1 Change Style

# Available Styles
1. darkgrid
2. whitegrid
3. dark
4. white
5. ticks

```python
sns.set_style('darkgrid')
sns.histplot(tips['total_bill'])
plt.show()
```

# 6.2 Change Color Palette

# Available Palettes
1. deep
2. muted
3. pastel
4. bright
5. dark
6. colorblind

```python
sns.set_palette('deep')
sns.countplot(x='day', data=tips)
plt.show()
```
